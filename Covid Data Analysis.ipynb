```python
satewise = pd.pivot_table(covid_df , values= ['Confirmed','Deaths','Cured'], index = "State/UnionTerritory", aggfunc = max)
```

```python
satewise
```

```python
satewise["Recovery Rate"] = satewise["Cured"] * 100 / satewise["Confirmed"]
```

```python
satewise["Mortality Rate"] = satewise["Deaths"] * 100 / satewise["Confirmed"] 
```

```python
satewise = satewise.sort_values(by="Confirmed" , ascending= False) 
```

```python
satewise.style.background_gradient(cmap= "cubehelix")
```

```python
#Top 10 States with most active cases in india
top_10_active_cases = covid_df.groupby(['State/UnionTerritory']).max()[['Active_Cases', 'Date']].sort_values(by = ['Active_Cases'], ascending=False).reset_index() 
sns.set(rc={'figure.figsize':(20,10)})
ax= sns.barplot(data=top_10_active_cases.iloc[:10] , x='State/UnionTerritory', y='Active_Cases', hue='State/UnionTerritory',legend=True)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Top 10 States with most active cases in india",size=20)
plt.show()
```

```python
#Top 10 States with highest deaths
top_10_deaths= covid_df.groupby(by= 'State/UnionTerritory').max()[["Deaths","Date"]].sort_values(by = 'Deaths', ascending=False).reset_index() 
fig = plt.figure(figsize=(18,5))
plt.title("Top 10 States With Most Deaths", size=25)
ax= sns.barplot(data=top_10_deaths.iloc[:12], y = "Deaths", x="State/UnionTerritory",hue="State/UnionTerritory", linewidth= 2, edgecolor='black')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
#Growth Trend
data=covid_df[covid_df['State/UnionTerritory'].isin(["Maharashtra","karnataka","Kerala","Tamil Nadu","Uttar Pradesh"])]
sns.set(rc={'figure.figsize':(20,10)})
ax= sns.lineplot(data=data , x = 'State/UnionTerritory' , y ='Active_Cases')
plt.show()
```
